```python
# Νίκος Λιθαρής ΑΜ:Π2019083
import re
import urllib.request
```

```python
# Step 1: Read web page text from input file
with urllib.request.urlopen("http://mixstef.github.io/courses/compilers/testpage.txt") as f:
    page_content = f.read().decode("utf-8")
```

```python
# Step 2: Export and print the title
title_pattern = re.compile("<title>(.*?)</title>")
title_match = title_pattern.search(page_content)
if title_match:
    title = title_match.group(1)
    print("Title: ", title)
```

```python
# Step 3: Delete comments
comment_pattern = re.compile("<!--.*?-->", re.DOTALL)
page_content = comment_pattern.sub(" ", page_content)
```

```python
# Step 4: Remove script and style tags
script_style_pattern = re.compile("<(script|style).*?>.*?</\\1>", re.DOTALL)
page_content = script_style_pattern.sub(" ", page_content)
```

```python
# Step 5: Extract and print links and link text
link_pattern = re.compile("<a.*?href=['\"](.*?)['\"].*?>(.*?)</a>")
for match in link_pattern.finditer(page_content):
    link = match.group(1)
    link_text = match.group(2)
    if link.startswith("http") or link.startswith("https"):
        print("Link: ", link)
        print("Link Text: ", link_text)
```

```python
# Step 6: Delete all tags
tag_pattern = re.compile("<.*?>")
page_content = tag_pattern.sub(" ", page_content)
```

```python
# Step 7: Convert HTML entities
entity_pattern = re.compile("&(amp|gt|lt|nbsp);")


def replace_entities(match):
    if match.group(1) == "amp":
        return "&"
    elif match.group(1) == "gt":
        return ">"
    elif match.group(1) == "lt":
        return "<"
    elif match.group(1) == "nbsp":
        return " "
    else:
        return match.group()


page_content = entity_pattern.sub(replace_entities, page_content)
```

```python
# Step 8: Convert sequences of consecutive whitespace characters to a single space character
whitespace_pattern = re.compile("\s+")
page_content = whitespace_pattern.sub(" ", page_content)
```

```python
# Step 9: Print the final formatted text
print(page_content)
```
